```python
def board(list1):
    print( '         |     |     ')
    print(f'      {list1[0]}  |  {list1[1]}  | {list1[2]} ')
    print('    ____ |____ |____ ')
    print('         |     |     ')
    print(f'      {list1[3]}  |  {list1[4]}  | {list1[5]} ')
    print('    ____ |____ |____ ')
    print('         |     |     ')
    print(f'      {list1[6]}  |  {list1[7]}  | {list1[8]} ') 
    print('         |     |     ')
```

```python
def choose():
    choice = False
    while choice == False:
        p = input('Enter X or O: ')
        if p == 'X' or p == 'O':
            if p == 'X':
                q = 'O'
                return [p ,q]
            else:
                q = 'X'
                return [p ,q]
            choice = True
        elif p == 'x' or p == 'o':
            if p == 'x':
                q = 'o'
                return [p ,q]
            else:
                q = 'x'
                return [p ,q]
            choice = True
        else:
            choice = False
```

```python
def player1(p,list1):
    choice = False
    while choice == False:
        x = input('Player1 enter the position (0 to 8): ')
        if x.isdecimal():
            x = int(x)
            if x in range(0,9):
                if x in [list1]:
                    choice = False
                else:
                    if list1[x]=='X' or list1[x]=='O':
                        print('Opps already entered')
                        choice = False
                    else:
                        list1[x] = p
                        board(list1)
                        choice = True
            else:
                print('Out of Range')
                ch = False
        else:
            print('Wrong Input!!!!!!')
            ch = False
    return list1
```
