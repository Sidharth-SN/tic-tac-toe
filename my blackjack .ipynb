```python
suits = ('diamonds','hearts','clubs','spades')
ranks = ('two','three','four','five','six','seven','eight','nine','ten','jack','queen','king','ace')
values = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9, 'ten':10, 'jack':10, 
          'queen':10, 'king':10, 'ace':11}
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
       
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
from random import shuffle
class Deck():
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
    def shuff(self):
        shuffle(self.all_cards)
        
    def remove(self):
        return self.all_cards.pop()
```

```python
class Bet():
    def __init__(self):
        self.money = 100
        self.bet_money = 0
    def bet(self):
        print(f'\nplayer money: {self.money}')
        a = True
        while a:
            try:
                self.bet_money = int(input('how much do u bet?  '))
                if self.bet_money <= self.money and self.bet_money > 0:
                    self.money -= self.bet_money
                    a = False
                    break
                else:
                    print('out of money!!')
                    print(f'remaning money: {self.money}')
                    a = True
            except:
                print('pls enter a number')
                a = True
                
    def win_bet(self):
        self.money += self.bet_money*2
    
    def lose_bet(self):
        self.money = self.money
        
    def draw_bet(self):
        self.money += self.bet_money
        
    def money1(self):
        return self.money
```

```python
class Dealer():
    def __init__(self):
        self.dealer_card=[]
        self.sum1 = 0
        
    def add(self,dcard):
        self.rank = dcard.split()[0]
        self.dealer_card.append(dcard)
        self.sum1 += values[self.rank]
        
        if self.rank =='ace' and self.sum1 > 21:
            print('ace is taken as 1 here')
            self.sum1 -= 10
            
    def sums(self):
        return self.sum1
        
    def display_some(self):
        print('\n DEALER CARDS:')
        print('<hidden card>')
        print(self.dealer_card[1])
        
    def display_all(self):
        print('\n DEALER CARDS:')
        print(*self.dealer_card, sep='\n')   
```
